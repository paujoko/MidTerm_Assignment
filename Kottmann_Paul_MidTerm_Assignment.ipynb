```python
import random
import time
import numpy
import matplotlib.pyplot as plt

def main_aufgabe_a():
    print('Aufgabe a:')
    print('I will now show my own matrix multiplication, multiplying two 8 dimentional square matrices:')
    matrix_1=first_matrix(8)
    matrix_2=second_matrix(8)
    print(f'First matrix:\n{matrix_1}')
    print(f'Second matrix:\n{matrix_2}')
    final_time, final_matrix=my_matrix_multiplication(8, matrix_2, matrix_1)
    print(f'The multiplication with my algorithm took {final_time} seconds.')
    print(f'This is my calculated matrix, resulting from the multiplication of the first and second matrices:\n{final_matrix} ')
    
def first_matrix(n):

    print('\n')
    print(f'The matrices 1 and 2 have {n} dimensionen:')
    matrix_1=[] #creates matrix 1
    for i in range(n):
        matrix_1_row=[] #sets new row of the Matrix to zero again
        for i in range(n): #creates every row of Matrix 1
            x= random.randint(0,9)
            matrix_1_row.append(x)
        matrix_1.append(matrix_1_row) #appends row to the original Matrix
    return matrix_1 

def second_matrix(n):

    matrix_2=[] #creates matrix 2 (same as matrix 1)
    for i in range(n):
        matrix_2_row=[]
        for i in range(n):
            x= random.randint(-100,100)
            matrix_2_row.append(x)
        matrix_2.append(matrix_2_row)
    return matrix_2


def my_matrix_multiplication(n, matrix_2, matrix_1): #performs my matrix multiplication
    
    start_time1= time.time()
    transpose_matrix_2=[] 
    for i in range(n):
        new_row= [row[i] for row in matrix_2] #puts all first elements together all second elements and so on to create a transpose of matrix_2
        transpose_matrix_2.append(new_row)

    final_matrix=[] #initialize the final matrix
    for i in range(n):
        final_matrix_row=[] #initialize a row for the final Matrix
        
        for k in range(n):
            final_matrix_element =0 #initialize the element to store in the row of the final matrix
            for y in range(n):
                final_matrix_element +=matrix_1[i][y]*transpose_matrix_2[k][y] #multiplies the elements of the matrix 1 and the transpose matrix at the same position  together and then adds it to the final element
            final_matrix_row.append(final_matrix_element)
        final_matrix.append(final_matrix_row)

    
    end_time1= time.time()
    final_time1= end_time1-start_time1 #measures time for the multiplication
    return final_time1, final_matrix

def numpy_multiplication(matrix_1,matrix_2):

    start_time2 = time.time()
    A = numpy.array(matrix_1)
    B = numpy.array(matrix_2)
    numpy.dot(A, B)  
    end_time2 = time.time()
    final_time2= end_time2-start_time2 #measures time for the multiplication
    return final_time2


main_aufgabe_a()
```

```python
def main_aufgabe_b():
    print('Aufgabe b:')
    my_time_list=[]
    for n in range(200,1100,200): #calculates two randomly generated matrices with the dimensions 200, 400, 600, 800, 1000 using my own matrix multiplication
        matrix_1=first_matrix(n)
        matrix_2=second_matrix(n)
        final_time_1, final_matrix=my_matrix_multiplication(n, matrix_2, matrix_1)
        print(f'The multiplication with my algorithm took {final_time_1} seconds.')
        my_time_list.append(final_time_1)
        
   
    

    dimensions=[]
    for n in range(200,1100,200):
        dimensions.append(n)    
    plt.plot(dimensions, my_time_list, label='My multiplication') #plots the time my multiplication needs depending on the number of dimensions
    plt.xlabel('Dimensions')
    plt.ylabel('Time in seconds')
    plt.title('My matrix multiplication')
    plt.legend()
    plt.show()
    

main_aufgabe_b()
```

```python
def main_aufgabe_c():
    print('Aufgabe c:')
    my_time_list=[]
    numpy_time_list=[]
    for n in range(200,1100,200):#calculates two randomly generated matrices with the dimensions 200, 400, 600, 800, 1000 using my own matrix multiplication algorithm and the Numpy algorithm.

        matrix_1=first_matrix(n)
        matrix_2=second_matrix(n)
        final_time_1, final_matrix=my_matrix_multiplication(n, matrix_2, matrix_1)
        print(f'The multiplication with my algorithm took {final_time_1} seconds.')
        my_time_list.append(final_time_1)
        final_time_2=numpy_multiplication(matrix_1,matrix_2)
        print(f'The multiplication with numpy took {final_time_2} seconds')
        numpy_time_list.append(final_time_2)
    
    dimensions=[]
    for n in range(200,1100,200):
        dimensions.append(n)    
    plt.plot(dimensions, my_time_list, label='My multiplication') #plots the Numpy multiplication and my own multiplication for comparison
    plt.plot(dimensions, numpy_time_list, label='Numpy multiplication')
    plt.xlabel('Dimensions')
    plt.ylabel('Time in seconds')
    plt.title('Comparison of the time it takes to multiply the matrices')
    plt.legend()
    plt.show()
    

main_aufgabe_c()
```
